 ## (a) Team Member A  pseudocode that describes the UBC algorithm is documented in the google doc/pdf of the report.

## (b) Team member B : Implementation of the UCB Algorithm using Team Member A’s pseudocode.

```python
import numpy as np

# Number of arms (stocks) and rounds (days)
num_stocks = len(daily_returns.columns)  # Number of different stocks (arms)
num_days = len(daily_returns)            # Number of trading days (rounds)

# Initialize arrays to track estimated values (means) and selection counts for each stock
estimated_values = np.zeros(num_stocks)  # Initialize estimated values for each stock to zero
selection_counts = np.zeros(num_stocks)  # Initialize selection counts for each stock to zero

# Initialize cumulative rewards tracker
ucb_cumulative_rewards = np.zeros(num_days)  # Array to hold cumulative rewards for each day

# Implement the Upper Confidence Bound (UCB) algorithm
for day in range(1, num_days):
    # Calculate UCB values for each stock
    ucb_values = estimated_values + np.sqrt((2 * np.log(day)) / (selection_counts + 1e-5))

    # Select the stock with the highest UCB value
    selected_stock = np.argmax(ucb_values)

    # Get the reward (return) for the selected stock on the current day
    reward = daily_returns.iloc[day, selected_stock]

    # Update the selection count for the selected stock
    selection_counts[selected_stock] += 1

    # Update the estimated value for the selected stock using incremental mean calculation
    estimated_values[selected_stock] += (reward - estimated_values[selected_stock]) / selection_counts[selected_stock]

    # Update cumulative rewards
    ucb_cumulative_rewards[day] = ucb_cumulative_rewards[day - 1] + reward

# Display the final selected stock and the estimated UCB values for all stocks
print("Final selected stock:", daily_returns.columns[selected_stock])
print("Final UCB values:", estimated_values)
```

## Team member C : Detailed comments on the UCB Algirithm




*   The code starts by importing the numpy library for array creation and
     manipulation.
*   **num_stocks**: Counts the number of different stocks, which are the "arms" in the multi-armed bandit problem.
*   **num_days**: Counts the number of trading days over which the algorithm will run.
*   **estimated_values**: Array to keep track of the estimated mean return for each stock.
*   **selection_counts**: Array to keep track of the number of times each stock has been selected.
*   **ucb_cumulative_rewards**: Array to track the cumulative rewards earned over the days.

*   We then have a loop that iterates thorugh every trading day from day 1
*   **ucb_values**: Calculated for each stock using the formula:


$$UCB_i = \mu_i + \sqrt{\frac{2log(t)}{n_i}}$$

*   $\mu_i$ is the estimated value of the stock
*   $t$ is the current day (time)

*   $n_i$ is the number of times the stock has been selected
*   **1e-5**: This is a small constant added to avoid division by zero when a stock has not been selected yet.

*  **selected_stock**: Index of the stock with the highest UCB value, indicating the stock to select for the current day.
*   **reward**: The actual return of the selected stock on the current day.

*   **selection_counts[selected_stock]**: Increments the count for the selected stock

*  **estimated_values[selected_stock]**: Updates the estimated mean return for the selected stock using the incremental mean formula.
*   **ucb_cumulative_rewards[day]**: Adds the reward from the current day to the cumulative total.


*   Lastly the output prints out the final selected stock and the estimated values

## **Team Member B**

 ## (a) Team Member B  pseudocode that describes the  Epsilon-greedy algorithm is documented in the google doc/pdf of the project report.

## **Team Member C**

## (b) Team Member C implements the steps of the Epsilon-Greedy Algorithm in Python using Team Member B’s pseudocode

```python

# Define the number of arms (stocks) and rounds (days)
n_arms = len(daily_returns.columns)  # Number of stocks
n_rounds = len(daily_returns)  # Number of days

# Epsilon-Greedy Parameters
epsilon = 0.1  # Epsilon value for exploration
eg_values = np.zeros(n_arms)  # Value estimates for each stock
eg_counts = np.zeros(n_arms)  # Count of selections for each stock
eg_cumulative_rewards = np.zeros(n_rounds)  # Cumulative rewards over time

# Run the Epsilon-Greedy algorithm
for t in range(1, n_rounds):
    if np.random.rand() < epsilon:
        # Exploration: randomly choose a stock
        selected_stock = np.random.choice(n_arms)
    else:
        # Exploitation: choose the stock with the highest estimated value
        selected_stock = np.argmax(eg_values)

    # Get the reward from the selected stock on day t
    reward = daily_returns.iloc[t, selected_stock]

    # Update the selection count and value estimate for the chosen stock
    eg_counts[selected_stock] += 1
    eg_values[selected_stock] += (reward - eg_values[selected_stock]) / eg_counts[selected_stock]

    # Update the cumulative rewards
    eg_cumulative_rewards[t] = eg_cumulative_rewards[t - 1] + reward

# Display the final selection and cumulative rewards
print("Final selected stock:", daily_returns.columns[np.argmax(eg_values)])
print("Final Epsilon-Greedy values:", eg_values)
```
